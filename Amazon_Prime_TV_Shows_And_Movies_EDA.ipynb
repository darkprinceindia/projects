## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

#custom function to parse list values
def parse_list_value(value):
    """Convert a stringified Python list into a clean list of values."""
    if pd.isna(value):
        return []
    if isinstance(value, list):
        return value
    try:
        parsed_value = ast.literal_eval(value)
        if isinstance(parsed_value, list):
            return [item for item in parsed_value if pd.notna(item)]
    except (ValueError, SyntaxError):
        return []
    return []

# Remove duplicated title ids and duplicated credit rows to avoid inflated counts.
titles = titles_raw.drop_duplicates(subset="id").copy()
credits = credits_raw.drop_duplicates().copy()

# Standardize missing descriptive and categorical values.
titles["description"] = titles["description"].fillna("No description available")
titles["age_certification"] = titles["age_certification"].fillna("Not Rated")


# Parse list-like columns.
titles["genres_list"] = titles["genres"].apply(parse_list_value)
titles["production_countries_list"] = titles["production_countries"].apply(parse_list_value)

# Create analysis-ready variables.
titles["primary_genre"] = titles["genres_list"].apply(lambda values: values[0] if values else "Unknown")
titles["main_country"] = titles["production_countries_list"].apply(lambda values: values[0] if values else "Unknown")
titles["genre_count"] = titles["genres_list"].apply(len)
titles["country_count"] = titles["production_countries_list"].apply(len)
titles["release_decade"] = (titles["release_year"] // 10 * 10).astype(int).astype(str) + "s"
titles["catalog_age"] = titles["release_year"].max() - titles["release_year"]

titles["runtime_bucket"] = pd.cut(
    titles["runtime"],
    bins=[0, 30, 60, 90, 120, 150, np.inf],
    labels=["0-30", "31-60", "61-90", "91-120", "121-150", "150+"],
    include_lowest=True,
)

titles["season_bucket"] = pd.cut(
    titles["seasons"],
    bins=[0, 1, 2, 3, 5, 10, np.inf],
    labels=["1", "2", "3", "4-5", "6-10", "10+"],
    include_lowest=True,
)

titles["imdb_votes_log"] = np.log10(titles["imdb_votes"].fillna(0) + 1)
titles["tmdb_popularity_log"] = np.log10(titles["tmdb_popularity"].fillna(0) + 1)

# Exploded helper datasets for genre and country level analysis.
genre_df = (
    titles[["id", "title", "type", "release_year", "imdb_score", "tmdb_score", "tmdb_popularity", "genres_list"]]
    .explode("genres_list")
    .rename(columns={"genres_list": "genre"})
)
genre_df["genre"] = genre_df["genre"].fillna("Unknown")

country_df = (
    titles[["id", "title", "type", "release_year", "imdb_score", "tmdb_score", "tmdb_popularity", "production_countries_list"]]
    .explode("production_countries_list")
    .rename(columns={"production_countries_list": "country"})
)
country_df["country"] = country_df["country"].fillna("Unknown")

# Summarize credits at title level and merge with title data.
credits_summary = credits.groupby("id").agg(
    total_people=("person_id", "nunique"),
    actor_count=("role", lambda values: (values == "ACTOR").sum()),
    director_count=("role", lambda values: (values == "DIRECTOR").sum()),
).reset_index()

eda_df = titles.merge(credits_summary, on="id", how="left")
eda_df[["total_people", "actor_count", "director_count"]] = eda_df[
    ["total_people", "actor_count", "director_count"]
].fillna(0)

numeric_features = [
    "release_year", "runtime", "seasons", "imdb_score", "imdb_votes",
    "tmdb_popularity", "tmdb_score", "genre_count", "country_count",
    "total_people", "actor_count", "director_count",
    "imdb_votes_log", "tmdb_popularity_log",
]

print("Analysis titles shape:", titles.shape)
print("Genre-level shape:", genre_df.shape)
print("Country-level shape:", country_df.shape)
print("Merged EDA shape:", eda_df.shape)
````

### What all manipulations have you done and insights you found?

I first removed duplicate ids from the titles dataset so that i can use it as unique identifier. Then I did the the same with credits dataset and made new dataset copies to work with. There were missing values in description and age certification, so i filled them with appropriate strings.

Early insights show that the catalogue is strongly movie-heavy, heavily concentrated in recent decades, and affected by incomplete age certification metadata.
